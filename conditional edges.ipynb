```python
from langgraph.graph import StateGraph
from typing import TypedDict, Dict
from langgraph.constants import START, END
```

```python
class AgentState(TypedDict):
    ops1: str
    ops2: str
    num1:int
    num2:int
    num3:int
    num4:int
    final_num1:int
    final_num2:int
```

```python
def add(state : AgentState)->AgentState:
    """greeet mfs who dont use toilet papers"""
    
    state["final_num1"] = state["num2"] + state["num1"]
    return state

def subtract(state:AgentState)->AgentState:
    state["final_num1"] = state["num2"] - state["num1"]
    return state

def add2(state : AgentState)->AgentState:
    """greeet mfs who dont use toilet papers"""
    
    state["final_num2"] = state["num2"] + state["num1"]
    return state

def subtract2(state:AgentState)->AgentState:
    state["final_num2"] = state["num2"] - state["num1"]
    return state
    

def router1(state : AgentState)->str:
    if state["ops1"] == "-":
        return "subtractor"
    if state["ops1"] == "+":
        return "adder"

def router2(state : AgentState)->str:
    if state["ops2"] == "-":
        return "subtractor2"
    if state["ops2"] == "+":
        return "adder2"
```

```python
graph = StateGraph(AgentState)
graph.add_node("adder",add)
graph.add_node("subtractor",subtract)
graph.add_node("adder2",add2)
graph.add_node("subtractor2",subtract2)
graph.add_node("router1",lambda state:state)
graph.add_node("router2",lambda state:state)
graph.add_edge(START,"router1")
graph.add_conditional_edges(
    "router1",
    router1,
    {
        "subtractor":"subtractor",
        "adder":"adder"
    }
    
)
graph.add_edge("adder","router2")
graph.add_edge("subtractor","router2")
graph.add_conditional_edges(
    "router2",
    router2,
    {
        "subtractor2":"subtractor2",
        "adder2":"adder2"
    }
    
)
graph.add_edge("adder",END)
graph.add_edge("subtractor",END)
app = graph.compile()

from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```
